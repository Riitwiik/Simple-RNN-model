```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb ##loading dataset from keras
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
## load imdb dataset
max_features=10000## Vocabulary Size
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=max_features)

## Print the shape of data
print(f'Training data shape:{X_train.shape}, Training labels shape: {y_train.shape}')
print(f'Testing data shape:{X_test.shape}, Testing labels shape: {y_test.shape}')
```

```python
sample_review=X_train[0]
sample_label=y_train[0]

print(f'Sample review (as integer):{sample_review}')
print(f'Sample label: {sample_label}')
```

```python
##Mapping of words index back to words(for Understanding)
word_index=imdb.get_word_index()
word_index
```

```python
#To print  index-word as value key pair
reverse_word_index={value:key for key,value in word_index.items()}
reverse_word_index
```

```python
## Decoding index to make them words if (i-3) does not exist in vocabulary replace by ? i is index of word
decoded_review=' '.join([reverse_word_index.get(i-3,'?') for i in sample_review])
decoded_review  ##So when you load word_index, the actual words start from index 1, but in the dataset they are stored starting from index 4
```

```python
from tensorflow.keras.preprocessing import sequence
max_len=500
X_train=sequence.pad_sequences(X_train,maxlen=max_len)
X_test=sequence.pad_sequences(X_test,maxlen=max_len)
X_train
```

```python
#Train Simple RNN

from tensorflow.keras import regularizers
from tensorflow.keras.layers import Dropout

model=Sequential()
model.add(Embedding(max_features,128,input_length=max_len))## Embedding layer
model.add(SimpleRNN(128,activation='tanh',kernel_regularizer=regularizers.l2(0.001))) #L2 regularization to fight overfitting
model.add(Dropout(0.5))#Dropout fight overfitting
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
##Create an instance of Early Stopping
from tensorflow.keras.callbacks import EarlyStopping
earlystopping=EarlyStopping(monitor='val_loss',patience=2,restore_best_weights=True)##restore best weight used for using best weights monitor=validation loss
earlystopping
```

```python
history=model.fit(X_train,y_train,epochs=10,batch_size=32,validation_split=0.2,callbacks=[earlystopping])
```

```python
#Save model file
model.save('simple_rnn_imdb.h5')
```
